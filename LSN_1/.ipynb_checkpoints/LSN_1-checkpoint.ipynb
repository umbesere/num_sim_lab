```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
import scipy.stats as stats
```

# ESERCIZIO 01

### ES01_1

Here's the graph of $<r>-\frac{1}{2}$ and its error. This value expresses the proximity of the calculated mean to the ideal value of $1/2$. It is calculated for an increasing number of blocks. It is clear that the function is compatible with zero as , for a high number of blocks, it lays slightly more than one sigma away from it.

```python
x, ave, err_ave, var, err_var = np.loadtxt("ES01_1.txt", usecols=(0,1,2,3,4), delimiter=' ', unpack='true', max_rows=100)
plt.errorbar(x,ave,yerr=err_ave)
plt.xlabel('#throws')
plt.ylabel('<r>-0.5')
plt.grid(True)
plt.show()
```

The figure shows $<(r-0.5)^2>-1/12$ vs the number of blocks used. The result is clear: for a high number of blocks involved the value is around $1\sigma$ from 0.

```python
plt.errorbar(x,var,yerr=err_var)
plt.xlabel('#throws')
plt.ylabel('$<(r-0.5)^2>-1/12$')
plt.grid(True)
plt.show()
```

```python
x, chi = np.loadtxt("ES01_1.txt", usecols=(0,1), delimiter=' ', unpack='true', skiprows=100)
plt.errorbar(x,chi)
plt.xlabel('#throws')
plt.ylabel('$\chi$')
plt.grid(True)
plt.show()
```

```python
unif1, exp1, cauchy1 = np.loadtxt("es01_2.txt", usecols=(0,1,2), delimiter=' ', unpack='true', max_rows=10000)
unif2, exp2, cauchy2 = np.loadtxt("es01_2.txt", usecols=(0,1,2), delimiter=' ', unpack='true', skiprows=10000)
unif3, exp3, cauchy3 = np.loadtxt("es01_2.txt", usecols=(0,1,2), delimiter=' ', unpack='true', skiprows=20000)
unif4, exp4, cauchy4 = np.loadtxt("es01_2.txt", usecols=(0,1,2), delimiter=' ', unpack='true', skiprows=30000)
unif = [unif1, unif2, unif3, unif4]
exp = [exp1, exp2, exp3, exp4]
cauchy = [cauchy1, cauchy2, cauchy3, cauchy4]
N = [1,2,10,100]
```

```python
fig, axx = plt.subplots(2, 2,figsize=(12, 12))
axx = axx.reshape(-1)
bins = np.linspace(0.,1.,100)

for k in range(4):
    ax = axx[k]
    ax.hist(unif[k], bins, density = True)
    ax.set_title(f"Uniform distribution with N = {N[k]}")
    ax.set_xlabel(r"x")
    ax.set_ylabel(r"dN/dx")
    ax.set_xlim(0.,1.)

mu, sigma = 1/2, 1/np.sqrt(12)/10
x = np.linspace(mu-3*sigma,mu+3*sigma,100)
ax.plot(x, stats.norm.pdf(x, mu, sigma), linewidth = 2)
```

```python
fig, axx = plt.subplots(2, 2,figsize=(12, 12))
axx = axx.reshape(-1)
bins = np.linspace(0.,5,100)

for k in range(4):
    ax = axx[k]
    ax.hist(exp[k], bins, density = True)
    ax.set_title(f"Exponential distribution with N = {N[k]}")
    ax.set_xlabel(r"x")
    ax.set_ylabel(r"dN/dx")
    #ax.set_xlim(0.,100.)

mu, sigma = 1., 1/10
x = np.linspace(mu-3*sigma,mu+3*sigma,100)
ax.plot(x, stats.norm.pdf(x, mu, sigma), linewidth = 2)
```

```python
fig, axx = plt.subplots(2, 2,figsize=(12, 12))
axx = axx.reshape(-1)
bins = np.linspace(-20.,20.,100)

for k in range(4):
    ax = axx[k]
    ax.hist(cauchy[k], bins, density = True)
    ax.set_title(f"Cauchy distribution with N = {N[k]}")
    ax.set_xlabel(r"x")
    ax.set_ylabel(r"dN/dx")
    #ax.set_xlim(0.,100.)

mu, width = 0.,1.
x = np.linspace(mu-5*width,mu+5*width,100)
ax.plot(x, stats.cauchy.pdf(x, mu, width), linewidth = 2)
```
