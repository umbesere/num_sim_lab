### EX 11_1

$$ f(x)=mx+b $$

As a first example, data sampled from a simple straight line will be used to train a single neuron with a linear activation function. Essentialy, this is equivalent to do a classic linear regression where the slope and the intercept are the weights resulting from the training of the neuron.

The training data will be generated adding gaussian noise to a stright line. Various cases with different amount of noise $\sigma$, training points $N_T$, validation points $N_V$ and epochs will be analised.

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

#### FIRST TRY

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 500)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.5 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

This is the graph representing the target (straight line with m=2 and b=1) and the validation data of the first try. In this case:
- $N_T = 500$
- $N_V = 50$
- $\sigma = 0.5$

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
# get a summary of our composed model
model.summary()
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
print()
print('Weights:')
# return weights
print('m = '+str(model.get_weights()[0][0][0]))
print('b = '+str(model.get_weights()[1][0]))
print()
print('Evaluating model:')
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
# print performance
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print()
print('Evaluating the model with the exact curve:')
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)
# print performance
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,4))

# look into training history
# Plot training & validation loss values
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.set_title('Model loss')
ax1.set_ylabel('Loss')
ax1.set_xlabel('Epoch')
ax1.legend(['Train', 'Test'], loc='best')
ax1.grid()

x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
ax2.scatter(x_predicted, y_predicted,color='r')
ax2.plot(x_valid, y_target)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid(True)
```

These two graphs show the loss in depence of the epochs and the comparison between the model prediction of points randomly generated in the fitting interval and the target function. They will be used along the notebook to evalute the quality of the fit.

In this case 30 epochs were sufficient to train the neuron, in fact the loss function of the training data becomes stable after around 15 epochs. The results are:
- m = 1.913626
- b = 0.9452583

They are not near the target as it can also be seen from the right hand graph. To try to reach better results one can rise the number of $N_T$ and/or the number of epochs.

### EX 11_2

$$
f(x)=4-3x-2x^2+3x^3
$$

In this second exercise the data are generated from the above polynomial with gaussian noise $\sigma=0.5$ in the interval $x\in (-1,1)$. This data will be used to train various netutral networks to see which one better approximates the function.

```python
def f(x):
    return 4 - 3*x-2*x**2+3*x**3
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 10000)
x_valid = np.random.uniform(-1, 1, 1000)
x_valid.sort()
y_target = [f(x) for x in x_valid] # ideal (target) linear functionprint
y_target = np.array(y_target)

sigma = 0.5 # noise standard deviation, for the moment it is absent
y_train = np.random.normal([f(x) for x in x_train], sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal([f(x) for x in x_valid], sigma)
```

```python
# plot training and target dataset
plt.plot(x_valid, y_target,label='target')
plt.scatter(x_valid, y_valid, color='r',label='validation')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True); plt.show()
```

Here's the graph illustrating the target function and the validation data. For each neural network the number of training points and valdiation points are fixed:
- $N_T = 10000$
- $N_V = 1000$

#### TWO-LAYER NETWORK

The first neural network used is a simple two-layer netowork. It is not a deep neural network and it was only useful to test the complexity necessary to obtain a good fit. It worked as a benchmark to build other networks. The activation function for the first layer is the hyperbolic tangent while the ouput layer has a simple linear activation function. The optimization function chosen in this case was the adam that, after some test, worked better than the sgd. Finally, the batch size remained fixed to 32 while the network eas trained using 50 epoch. Finally and most importantly, the first layer consisted in 100 neurons.

```python
model = tf.keras.Sequential()
model.add(Dense(100, input_shape=(1,), activation='tanh'))
model.add(Dense(1, activation='linear'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='adam', loss='mse', metrics=['mse'])


# get a summary of our composed model
model.summary()

# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
#callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10, baseline=5, restore_best_weights=True)

history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=50,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# look into training history
# Plot training & validation loss values
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(16,4))
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.set_title('Model loss')
ax1.set_ylabel('Loss')
ax1.set_xlabel('Epoch')
ax1.legend(['Train', 'Test'], loc='best')
#plt.show()

x_predicted = np.random.uniform(-1.5, 1.5, 500)
y_predicted = model.predict(x_predicted)
ax2.scatter(x_predicted, y_predicted,color='r', label='model')
#plt.scatter(x_valid, y_valid,color='b')
x_predicted.sort()
y_tar = [f(x) for x in x_predicted]
ax2.plot(x_predicted, y_tar, label='target funct')
ax2.grid(True)
ax2.set_title('Model prediction')
ax2.vlines([-1,1],0,5, label='training interval')
ax2.set_xlim(-2,2)
ax2.set_ylim(0,5)
ax2.legend()
#plt.show()
```

In this case the networks work well in the central region of the interval. It does not manage to fit the regions arond x=-1 and x=1 (inside the fitting interval) properly. It is not a bad fit in general but, using more layers, the hope is to obtain a good results in the extremities of the interval, too.\
In the region outside the fitting interval the model cannot be trusted as it can be seen in the right hand graph. The model prediction moves away from the target as x moves away from x=-1 or x=1. 

 #### FOUR-LAYER NETWORK

This time a four-layer network is trained to fit the data. This network consists in in a 75-neurons layer with a sigmoid AF, two 20-neurons layers  with an elu AF and a final one neuron layer with elu AF. In general, sigmoids and elus work better than relus, especially in small networks, because they are smooth. The adam was chosen as OF over the sgd because of better results while 100 epochs were needed to train the model in batches of size 32. The loss function used was the mse as in the one-layer network.

```python
model = tf.keras.Sequential()
model.add(Dense(75, input_shape=(1,), activation='sigmoid'))
model.add(Dense(20, input_shape=(1,), activation='elu'))
model.add(Dense(20, input_shape=(1,), activation='elu'))
model.add(Dense(1, activation='elu'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='adam', loss='mse', metrics=['mse'])


# get a summary of our composed model
model.summary()

# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
#callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10, baseline=5, restore_best_weights=True)

history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# look into training history
# Plot training & validation loss values
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(16,4))
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.set_title('Model loss')
ax1.set_ylabel('Loss')
ax1.set_xlabel('Epoch')
ax1.legend(['Train', 'Test'], loc='best')
#plt.show()

x_predicted = np.random.uniform(-1.5, 1.5, 500)
y_predicted = model.predict(x_predicted)
ax2.scatter(x_predicted, y_predicted,color='r', label='model')
#plt.scatter(x_valid, y_valid,color='b')
x_predicted.sort()
y_tar = [f(x) for x in x_predicted]
ax2.plot(x_predicted, y_tar, label='target funct')
ax2.grid(True)
ax2.set_title('Model prediction')
ax2.vlines([-1,1],0,5, label='training interval')
ax2.set_xlim(-1.5,1.5)
ax2.set_ylim(0,5)
ax2.legend()
#plt.show()
```

This time the networks manages to approximate the function well in the entire interval (-1,1) but, as in the one-layer network, the model is not good for prediction outside this interval.

#### SIX LAYERS NETWORK

The last network tried is a six-layer netowrk consisted in: a 16-neurons layer with an hyperbolic tangent as AF, a 20-neurons layer with elu as AF, two 12-neurons layers with elu as AF and a final layer with one neuron with elu as AF. Sigmoid is similar to tanh and can be used in the first node instead of the latter. Adam is used as AF, the loss function is the mse and, once again, 100 epochs with batches of size 32 are used to train the network.

```python
model = tf.keras.Sequential()
model.add(Dense(24, input_shape=(1,), activation='tanh'))
model.add(Dense(12, activation='elu'))
model.add(Dense(9, activation='elu'))
model.add(Dense(10, activation='elu'))
model.add(Dense(15, activation='elu'))
model.add(Dense(1, activation='elu'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='adam', loss='mse', metrics=['mse'])


# get a summary of our composed model
model.summary()

# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
#callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10, baseline=5, restore_best_weights=True)

history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# look into training history
# Plot training & validation loss values
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(16,4))
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.set_title('Model loss')
ax1.set_ylabel('Loss')
ax1.set_xlabel('Epoch')
ax1.legend(['Train', 'Test'], loc='best')
#plt.show()

x_predicted = np.random.uniform(-1.5, 1.5, 500)
y_predicted = model.predict(x_predicted)
ax2.scatter(x_predicted, y_predicted,color='r', label='model')
#plt.scatter(x_valid, y_valid,color='b')
x_predicted.sort()
y_tar = [f(x) for x in x_predicted]
ax2.plot(x_predicted, y_tar, label='target funct')
ax2.grid(True)
ax2.set_title('Model prediction')
ax2.vlines([-1,1],0,5, label='training interval')
ax2.set_xlim(-1.5,1.5)
ax2.set_ylim(0,5)
ax2.legend()
#plt.show()
```

Even if the network is not as wide as the others, it can fit the polynomial pretty well in the fitting interval. As the others two, it fails outside this interval.

#### FINAL CONSIDERATIONS ON EX 11_2

The first thing to observe about these networks is that, even if the last two are different in terms of complexity (4 and 6 layers) and parameters (2111 and 746), they can both fit the polynomial inside the interval. Obviously, from a computational point of view, it is more convienent to choose the six-layer network because it has much less parameters. This tells us that increasing the complexity of a network is not always equal to increase the goodness of the fit (here the more complex is only slightly better the the other one). In general, it also depends on other factors like the number of training points. \
From the results above, it can also be said that neural networks do not have any predictive power outside the training interval.

### EX 11_3

$$ f(x,y)=\sin ( x^2 + y^2 ) $$

In this final exercise, data are generated from the above function adding gaussian noise with $\sigma = 0.5$.
in this case the number of training and validation points were chosen to be $N_T=10000$ and $N_V=1000$. Obviously, this time the x data are generated inside a square.

```python
def f(x,y):
    return np.sin(x**2+y**2)
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-3./2., 3./2., size=(10000,2))
x_valid = np.random.uniform(-3./2., 3./2., size=(1000,2))

y_target = [f(x,y) for (x,y) in x_valid] # ideal (target) linear functionprint
#print(len(y_target))
y_target = np.array(y_target)
#print(y_target)
sigma = 0.5 # noise standard deviation, for the moment it is absent
y_train = np.random.normal([f(x,y) for (x,y) in x_train], sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal([f(x,y) for (x,y) in x_valid], sigma)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection='3d')

ax.scatter(x_valid[:,0], x_valid[:,1], y_valid, c='black', marker='.', s=100)
X = np.arange(-3/2,3/2,0.1)
Y = np.arange(-3/2,3/2,0.1)
X, Y = np.meshgrid(X,Y)
Z = np.sin(X**2+Y**2)
ax.plot_wireframe(X, Y, Z)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
#ax.view_init(10, 30)
```

This graph shows the validation points and the target mesh.

The network chosen is a relatively simple five layers network with a first 12-neurons layer in tanh as AF, a second layer with 16 neurons with a tanh AF, two layers with 8 neurons and a tanh as AF and a final linear neuron. Obviously, the neurons in the first layer have two inputs because the x data are 2D. In spite of the past newtoworks, in this one the plateau AF worked better then the unbounded ones. \
The networked wa trained for 200 epochs with batches of size 64. The stochastic gradient descend was chosen over the adam while the usual mean square error was used as loss function.

```python
model = tf.keras.Sequential()
model.add(Dense(12, input_shape=(2,), activation='tanh'))
model.add(Dense(16, activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(1, activation='linear'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()

# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=64, epochs=200,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))

# return weights
model.get_weights()

# evaluate model
print('Evaluate the model:')
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
print('Evaluate the model with the exact curve:')
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)
# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

x_predicted = np.random.uniform(-3/2, 3/2, size=(1000,2))
y_predicted = model.predict(x_predicted)
y_predicted = [x[0] for x in y_predicted]

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection='3d')
ax.scatter(x_predicted[:,0], x_predicted[:,1], y_predicted, c='black', marker='.', s=100,label='prediction')
X = np.arange(-3/2,3/2,0.1)
Y = np.arange(-3/2,3/2,0.1)
X, Y = np.meshgrid(X,Y)
Z = np.sin(X**2+Y**2)
ax.plot_wireframe(X, Y, Z, label='target')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
ax.set_title('Model prediction')
ax.legend()
plt.show()
```

The first graph shows a significant reduction of the loss function durring the training. \
The second one shows that the prediction of the model is very close to the mesh representing the target.
