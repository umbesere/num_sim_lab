### ES04_1

```python
states = ["solid","liquid"]
fig, axx = plt.subplots(2, 1,figsize=(12, 12))
axx = axx.reshape(-1)
for j in range(2):
    etot = np.loadtxt("equilibration/output_etot_"+states[j]+".dat", usecols = (0))
    epot = np.loadtxt("equilibration/output_epot_"+states[j]+".dat", usecols = (0))
    ekin = np.loadtxt("equilibration/output_ekin_"+states[j]+".dat", usecols = (0))
    temp = np.loadtxt("equilibration/output_temp_"+states[j]+".dat", usecols = (0))

    measures = [etot, epot, ekin, temp]
    labels = ["etot", "epot", "ekin", "temp"]

    x = np.linspace(0, len(etot), len(etot))

    ax = axx[j]
    for i, mea in enumerate(measures):
        ax.errorbar(x, mea, label = labels[i])
    ax.set_title(states[j]+"_equilibration")
    ax.set_xlabel(f"steps")
    ax.set_ylabel(f"phases in reduced units")
    ax.grid(True)
    ax.legend(loc=(0.8, 0.55))
```

The figures above show the equilibration for the stadard input.liquid and input.solid files. 5 cycles of the algorithm are represented here. It is clear that, while in the first couple of cycles the phases abruptly shift at the beginning, once the equilibrium is reached (4th - 5th cycle) the transition between two cycles is almost invisible. Moreover, in the end of the equilibration, the temperature is compatible with the one forced at the beginning of the first cycle.

To implement the rescaling of the velocities and the possibility to restart, two boolean variables were added to the set of parameters to start a simulation. One expresses the possibility to restart from an old configuration while the second one is useful to rescale velocities during the equilibration phase.

### ES04_3

```python
sigma = 0.34e-9
eps_over_kb = 120
m = 39.948*1.66e-27
kb = 1.38e-23

# 5 cycles to equilibrate
fs = [eps_over_kb*kb, eps_over_kb*kb, eps_over_kb*kb, eps_over_kb]
#fs = [1,1,1,1]
time_fs = 0.1*sigma*np.sqrt(m/eps_over_kb/kb)
#time_fs = 1
units = ["J", "J", "J", "K"]
state = ["solid","liquid","gas"]
labels = ["etot", "epot", "ekin", "temp"]
fig, axx = plt.subplots(3, 4,figsize=(27,16) )
axx = axx.reshape(-1)
for j in range(len(state)):
    for i in range(4):
        en, en_err = np.loadtxt(state[j]+"_argon/average_"+labels[i]+".out", unpack = True, usecols = (2,3))
        #en, en_err = np.loadtxt("gas_argon/average_"+labels[i]+".out", unpack = True, usecols = (2,3))

        x = np.linspace(0, len(en), len(en_err))

        ax = axx[4*j+i]
        ax.errorbar(x*time_fs, en*fs[i], en_err*fs[i])
        ax.set_title(state[j]+"_"+labels[i]+"_average")
        ax.set_xlabel(f"time[s]")
        ax.set_ylabel(labels[i]+f"_average[{units[i]}]")
        ax.grid(True)
```

```python
temp = [0.8, 1.1, 1.2]
for i in range(3):
    print(state[i]+" argon ideal temp: "+str(temp[i]*eps_over_kb)+" K\n")
```

In this set of graphs all the progressive means of interest are represented for solid, liquid and gas argon. All the quantities are expressed in SI units (one can change to reduced units uncommenting two rows of the code) and the progressive means are calculated with the method described above.

Solid argon is the most stable of the three while gas argon is the most unstable. In my opinion, it is simply because gas argon is much more correlated than solid or liquid so it fluctuates more with a fixed number of moves per block.
