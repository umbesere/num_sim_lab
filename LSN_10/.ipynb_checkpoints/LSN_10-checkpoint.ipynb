# LSN_10

In this notebook the travelling salesman problem is solved once again with two new methods. The first one is the simulated annealing while the second is an application of parallel computing to genetic algorithms.

### ES10_1

The idea behind the simulating annealing is that if the cost/loss function is interpreted as the energy (hamiltonian) of a system, then one can reach the minum of the energy (lowest cost) slowly cooling the system while sampling the Boltzmann weight. The concept is similar to the phase transition between liquid and solid. If a liquid is slowly cooled, it will have a characteristic crystalline structure (lowest energy configuration) once the solid state is reached. The same does not happen when the liquid is abruptly cooled: the system will find itself in a metastable solid state (the configuration will not be the lowest in energy). From the point of view of optimization, slowly cooling the system while sampling the Boltzmann weight will take the system to the minimum of the cost function (energy).

Practically, let's look at how to solve the TSP for 32 cities on a plan. First, the system has to start from a random path at a sufficiently high temperature. In this case, the starting temperature chosen is 10.001. Then, the boltzmann weight is sampled with the Metropolis algorithm exploiting the mutations of the genetic algorithm in NSL_9. In detail, the new mutated configuration is computed and, if the length of the new path is lower than the old one, the new one replaces the old one. Otherwise, the path is replaced by the mutated one with probability $e^{-\beta \Delta L}$. The sampling is of the Gibbs type so no accepatance rate has to be checked. The mutation operators used are the same of NSL_9:
- swap operator: it picks two cities at random and swap them
- shift operator: it picks a random set of contiguous cities and shift them of a random amount
- continuous swap operator: it picks two random non-intersecting set of contiguous cities with equal length and swap them
- inversion operator: it picks a random set of contiguous cities and invert the order in which they are visited

These mutations are tried for a sufficiently high number of times, in this case 100. Finally, the temperature is reduced of a fixed amount. The amount chosen is 0.001.

These steps are repeated until the temperature is low enough that the system has reached (or is near to ) the minimum of the energy/loss function. In these case the steps are repeated for $10^6$ times reaching the temperature of 0.001.

The most important thing to be aware of is that the system must be cooled very slowly to guarantee that the final path is (or is close to) the optimal one.

#### Circumference

First the TSP is solved for 32 cities on a circumference to test the simulated annealing.

```python
niter = np.loadtxt("Circle/es10_1.txt", usecols=(0), skiprows=1, max_rows=1, unpack='true')
niter = int(niter)
step, best = np.loadtxt("Circle/es10_1.txt", usecols=(0,1), skiprows=3, max_rows=niter, unpack='true')
best_path = np.loadtxt("Circle/es10_1.txt", skiprows=niter+3, max_rows=1)
x_city, y_city = np.loadtxt("Circle/map.txt", usecols=(0,1), unpack='True')
```

```python
fig, axx = plt.subplots(2, 1,figsize=(12, 12))
axx.reshape(-1)

ax = axx[0]
ax.plot(step, best, label = 'best_path')
for i in range(11):
    ax.axvline(i*100000, color = 'red')
ax.set_title("PATH LENGTH vs GENERATION")
ax.set_xlabel(f"#steps")
ax.set_ylabel(f"length")
ax.legend()
ax.grid(True)
print('Best route = '+str(best_path))
print('Best route length = '+str(best[-1]))

ax = axx[1]
start = 950000
end = 1000000
ax.errorbar(step[start:end], best[start:end], label = 'best_path')
#for i in range(3):
#    ax.axvline(800000+i*100000, color = 'red')
ax.set_title("PAATH LENGTH vs GENERATION (150 to 500)")
ax.set_xlabel(f"#steps")
ax.set_ylabel(f"length")
ax.legend()
ax.grid(True)
```

In the first graph the path length is represented as a function of the number of steps. The vertical red bars indicate every time the temperature is decreased by a unit. As previously stated, they start from 10.001 at step 0 and end at 0.001 at step $10^6$. Every step corresponds to a try of the four mutations described above (10000 temperatures times 100 tries of the four mutations for each temperature).

It is interesting to see that the oscillations of the length are progressively reduced to essentialy zero as the system cools down. In particular, it is well noticeable in the second plot where the last 50000 steps are presented.

```python
x_path = [ x_city[int(k)] for k in best_path ]
y_path = [ y_city[int(k)] for k in best_path ]

fig, ax = plt.subplots(1, 1,figsize=(8, 8))
#ax.plot(x_path, y_path, marker = 's')
for i in range(len(x_path)-1):
    ax.arrow(x_path[i],y_path[i],x_path[i+1]-x_path[i],y_path[i+1]-y_path[i],head_width=0.03, head_length=0.025, length_includes_head='yes')
ax.arrow(x_path[-1],y_path[-1],x_path[0]-x_path[-1],y_path[0]-y_path[-1],head_width=0.03, head_length=0.025,length_includes_head='yes')
ax.set_title("BEST PATH")
ax.set_xlabel(f"X")
ax.set_ylabel(f"Y")
ax.set_xlim(-1.2,1.2)
ax.set_ylim(-1.2,1.2)
ax.grid(True)
```

This is a plot of the best path reached by the simulated annealing. For the circumference is quite simple to understand that this is the optimal solution of the TSP. The length of this path can be read above the previous graph (1.91674).

### ES10_2

The second method used to solve the TSP is a genetic algorithm upgraded with the use of parallel computing. In this case, the parallel computing is exploited to carry out multiple computations simultaneously. In detail, a single program will be executed at the same time on a fixed number of cores.

But how can parallel computing work well with genetic algorithms? The suggestion comes from Nature once again. Let's think as the processes carried out at the same time, say 4, as continents. Each continent has its own population and they can exchange (through the MPI) some induviduals every fixed number of generations. This will guarantee more variability on the whole and hence a better approximation of the optimal solution of TSP. The MPI is a standard in parallel computing that permits to write a program executable at the same time by multiple processors.

Now, let's go deeper inside the solution of TSP. The GA structure used is the same of NSL_9(4 mutations + crossover, 500 individuals, 1000 generations) but this time will be carried out on 4 cores. The objective is to translate into code the idea shown in the last paragraph. First of all, one have to make sure different cores start to generate random numbers from different seeds, otherwise the 4 cores will have the same story. Another thing to notice is that every process must use the same map (position of the 32 cities) so that they will solve the same problem. In this case, the same map of NSL_9 is used in order to compare directly the two methods. \
The trickiest part is to define how the 4 processes exchange individuals. This issue has been solved this way. Every 10 generations, the 20 fittest individuals (shortest paths) of each population are exchanged so that process 1 passes its "migrants" to process 2, process 2 to process 3, process 3 to process 4 and process 4 back to process 1.($*$)
$$ 1 \rightarrow 2,\quad 2 \rightarrow 3,\quad 3 \rightarrow 4,\quad 4 \rightarrow 1$$
The number of "migrants" and the migrations frequency were chosen after some oculate testing (high parameters cause too much variability during time). \
To pass individuals between populations, MPI_SEND and MPI_RECV were exploited. In particular, every process execute a loop that guarantees no deadlocks.

($*$) Another solution can be to exchange the migrants randomly. For what I've seen, the solution path is longer and more unstable (check square_PCGA_rand).

These four graphs serve as comparison between GA and parallel GA.\
The first two show the length of the best path of the population at a given generation. In particular, the first one illustrates the lengths of the best path in all the four populations. Comparing the first two figures, it can be seen that the two methods have a similar convergence time in terms of number of steps (around 400-500) but the parallel GA is able to reach lower lengths. \
The last two graphs are a focus of the first two between generation 200 and 600. The four populations of the parallel GA converge almost simultaneously and definetly to the value of 3.99 while the GA is not stable until generation ~450 where there is a little step in the length and it reaches 4.32543.

Overall, the parallel GA reaches lower lengths and it can be considered a better solver for the TSP.

NB: the computational time to solve the problem with parallel GA is higher. This is due to the MPI_SEND-MPI_RECV loop and a MPI_GATHER that forces the four processes to stop and wait for the others to reach the line. Using MPI_IRECV does not decrease sensibly the preformance.

This is a comparison between the best paths computed with parallel GA and GA. The right hand side of the path changes sensibly from the parallel to the non parallel. It clearly shows that the result of NSL_9 was only a fair approximation of the true local minimum that is not reached even by the parallel GA as it is shown in the next graph.

```python
fig, ax = plt.subplots(1, 1,figsize=(12, 6))
start = 400
end = 1000
ax.errorbar(step0[start:end], best0[start:end], label = 'best_path_rank0')
ax.errorbar(step1[start:end], best1[start:end], label = 'best_path_rank1')
ax.errorbar(step2[start:end], best2[start:end], label = 'best_path_rank2')
ax.errorbar(step3[start:end], best3[start:end], label = 'best_path_rank3')
ax.set_title("PATH LENGTH vs GENERATION ("+str(start)+" to "+str(end)+" generations), GA")
ax.set_xlabel(f"generation")
ax.set_ylabel(f"length")
ax.legend()
ax.grid(True)
```

This last graph shows that the minimum reached by the parallel GA is only local. It illustrates the length as a function of the generation number from generation 400 to 1000. There are times when a population manage to arrive at a new minimum lower than 3.99 then, probably because of mutations/migrations or because the new minimum is shallow, it returns to 3.99.

#### CONCLUSION

In conclusion, these methods of random search are very useful to approach a local minuimum that fairly approximates the global one in complex problems like the TSP with 32 cities in a square. It is also true that reaching the global minimum is a very difficult task but, in most pracitcal cases, not so useful.
